#  ğ—”ğ˜€ğ—¸ ğ˜ğ—µğ—² ğ˜‚ğ˜€ğ—²ğ—¿ ğ˜ğ—¼ ğ—²ğ—»ğ˜ğ—²ğ—¿ ğ—® ğ—»ğ˜‚ğ—ºğ—¯ğ—²ğ—¿. ğ—–ğ—µğ—²ğ—°ğ—¸ ğ—¶ğ—³ ğ—¶ğ˜'ğ˜€ ğ—±ğ—¶ğ˜ƒğ—¶ğ˜€ğ—¶ğ—¯ğ—¹ğ—² ğ—¯ğ˜† ğŸ± ğ—¼ğ—¿ ğŸ³, ğ—¼ğ—¿ ğ—¯ğ—¼ğ˜ğ—µ, ğ—®ğ—»ğ—± ğ—½ğ—¿ğ—¶ğ—»ğ˜ ğ—® ğ—ºğ—²ğ˜€ğ˜€ğ—®ğ—´ğ—².

```python
num = 15

if num % 5 == 0 and num % 7 == 0:
    print("Divisible by both 5 and 7")
elif num % 5 == 0:
    print("Divisible by 5")
elif num % 7 == 0:
    print("Divisible by 7")
else:
    print("Not divisible by 5 or 7")
```
